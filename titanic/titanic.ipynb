```python
df=pd.read_csv("Titanic-Dataset.csv")
```

```python
df=df[['Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
```

```python
df.groupby('Pclass').agg(
    survival_rate=('Survived', 'mean'),
    avg_fare=('Fare', 'mean'),
    avg_age=('Age', 'mean')
).round(2)
```
