```python
import time
```

```python

size=10_000_000

python_list=list(range(size))
start=time.time()

list_squared=[x**2 for x in python_list]
end=time.time()

print("Python list time ", end-start, "seconds")

np_array=np.array(python_list)
start=time.time()

array_squared=np_array **2
end=time.time()

print("Numpy array time ", end-start, "seconds")
```
