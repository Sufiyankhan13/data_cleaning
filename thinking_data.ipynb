```python
def clean_data(orders):
    for order in orders:

        # --- Clean Name ---
        name = order["customer"]["name"]
        order["customer"]["name"] = name.strip().title()

        # --- Clean Email ---
        email = order["customer"]["email"]
        if email:
            email = email.strip().lower()
            if "@" not in email:
                email = None
            order["customer"]["email"] = email

        # --- Clean Age ---
        age = order["customer"]["age"]
        if age is None or not str(age).isdigit() or int(str(age)) > 100:
            order["customer"]["age"] = None
        else:
            order["customer"]["age"] = int(age)

    return orders
            
```

```python
def display_data(orders):
    for order in orders:
        print(f"Order: {order['order_id']} | Name: {order['customer']['name']} | Email: {order['customer']['email']} | Age: {order['customer']['age']}")
```

```python
def clean_items(orders):
    for order in orders:
        for item in order["items"]:

        # clean Name

            name = item["name"]
            item["name"]=name.strip().title()

            # Clean the Category 
            category = item["category"]
            item["category"]=category.strip().title()

        # --- Clean Product ID ---
            pid = item["product_id"]
            item["product_id"] = pid.strip().upper()

            # --- Clean Price ---
            price = item["price"]
            if isinstance(price, str):
                item["price"] = float(price.strip())
            elif price < 0:
                item["price"] = None  # negative price = invalid

            # --- Clean Quantity ---
            quantity = item["quantity"]
            if isinstance(quantity, str):
                item["quantity"] = int(quantity.strip())
            elif quantity == 0:
                item["quantity"] = None  # zero quantity = invalid

    return orders

        
```

# Cleaning the Duplicates



```python
def inspect_duplicates(orders):
    
    # Check duplicate order IDs
    seen_ids = []
    for order in orders:
        oid = order["order_id"]
        if oid in seen_ids:
            print(f"  [DUPLICATE ORDER ID] {oid}")
        else:
            seen_ids.append(oid)

    # Check duplicate items inside each order
    for order in orders:
        seen_items = []
        for item in order["items"]:
            pid = item["product_id"]
            if pid in seen_items:
                print(f"  [DUPLICATE ITEM] {pid} in order {order['order_id']}")
            else:
                seen_items.append(pid)

# Run it
inspect_duplicates(orders)
```

```python
def remove_duplicates(orders):

    # Remove duplicate orders
    seen_ids = []
    cleaned_orders = []
    for order in orders:
        oid = order["order_id"]
        if oid not in seen_ids:
            seen_ids.append(oid)
            cleaned_orders.append(order)
        else:
            print(f"  [REMOVED] duplicate order {oid}")

    # Remove duplicate items inside each order
    for order in cleaned_orders:
        seen_items = []
        unique_items = []
        for item in order["items"]:
            pid = item["product_id"]
            if pid not in seen_items:
                seen_items.append(pid)
                unique_items.append(item)
            else:
                print(f"  [REMOVED] duplicate item {pid} in order {order['order_id']}")
        order["items"] = unique_items

    return cleaned_orders

# Run it
orders = remove_duplicates(orders)
```
